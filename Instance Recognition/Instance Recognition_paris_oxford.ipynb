```python
from torchvision import transforms
from PIL import Image
import torch
import torch.nn.functional as F
import os


def get_preprocessing_pipeline():
    return transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])


def load_and_preprocess_image(image_path, preprocess_pipeline):
    image = Image.open(image_path).convert("RGB")
    return preprocess_pipeline(image).unsqueeze(0)


def extract_image_features(model, image_tensor):
    with torch.no_grad():
        return model(image_tensor)


def compute_cosine_similarity(query_features, database_features_list):
    similarities = [F.cosine_similarity(query_features, db_features) for db_features in database_features_list]
    return similarities


def rank_similarities(similarities):
    sorted_indices = torch.argsort(torch.tensor(similarities), descending=True)
    return sorted_indices
```

```python
import os
import torch
from PIL import Image
import torch.nn.functional as F
import numpy as np


def load_query_good_ok_junk(groundtruth_dir, group_name):
    query_file = os.path.join(groundtruth_dir, f'{group_name}_query.txt')
    good_file = os.path.join(groundtruth_dir, f'{group_name}_good.txt')
    ok_file = os.path.join(groundtruth_dir, f'{group_name}_ok.txt')
    junk_file = os.path.join(groundtruth_dir, f'{group_name}_junk.txt')

    query_image = load_image_list(query_file)[0]
    good_images = load_image_list(good_file)
    ok_images = load_image_list(ok_file)
    junk_images = load_image_list(junk_file)

    return query_image, good_images, ok_images, junk_images


def load_image_list(file_path):
    with open(file_path, 'r') as f:
        image_list = []
        for line in f:
            filename = line.split()[0] + ".jpg"
            image_list.append(filename)
    return image_list


def find_images_in_dataset(dataset_dir, image_filenames):
    image_paths = []
    for root, dirs, files in os.walk(dataset_dir):
        for filename in image_filenames:
            if filename in files:
                image_paths.append(os.path.join(root, filename))
    return image_paths


def load_and_preprocess_image(image_path, preprocess_pipeline):
    image = Image.open(image_path).convert("RGB")
    return preprocess_pipeline(image).unsqueeze(0)


def extract_image_features(model, image_tensor):
    with torch.no_grad():
        return model(image_tensor)


def compute_cosine_similarity(query_features, db_features_list):
    similarities = [F.cosine_similarity(query_features, db_features, dim=1).item() for db_features in db_features_list]
    return similarities


def compute_mAP(similarities, relevant_images, total_relevant):
    sorted_indices = np.argsort(similarities)[::-1]
    retrieved = 0.0
    precision = 0.0
    relevant_count = 0.0
    ap = 0.0

    for i, idx in enumerate(sorted_indices):

        if relevant_images[idx] > 0:
            relevant_count += relevant_images[idx]
            retrieved += 1
            precision = relevant_count / (retrieved + 0.0)
            ap += precision * relevant_images[idx]

    return ap / total_relevant if total_relevant > 0 else 0



def process_ground_truth(groundtruth_dir, dataset_dir, model, preprocess_pipeline):

    group_names = extract_group_names(groundtruth_dir)

    total_mAP = 0.0
    total_H = 0.0
    num_queries = 0

    for group_name in group_names:
        query_image_name, good_image_names, ok_image_names, junk_image_names = load_query_good_ok_junk(groundtruth_dir, group_name)

        if 'oxford' in dataset_dir:
            query_image_name = query_image_name.replace('oxc1_', '')
            good_image_names = [name.replace('oxc1_', '') for name in good_image_names]
            ok_image_names = [name.replace('oxc1_', '') for name in ok_image_names]
            junk_image_names = [name.replace('oxc1_', '') for name in junk_image_names]



        query_image_path_list = find_images_in_dataset(dataset_dir, [query_image_name])
        if not query_image_path_list:
            print(f"Query image not found in dataset for group: {group_name}, image name: {query_image_name}")
            continue
        query_image_path = query_image_path_list[0]


        good_image_paths = find_images_in_dataset(dataset_dir, good_image_names)
        ok_image_paths = find_images_in_dataset(dataset_dir, ok_image_names)
        junk_image_paths = find_images_in_dataset(dataset_dir, junk_image_names)


        query_image_tensor = load_and_preprocess_image(query_image_path, preprocess_pipeline)
        query_features = extract_image_features(model, query_image_tensor)


        good_tensors = [load_and_preprocess_image(img, preprocess_pipeline) for img in good_image_paths]
        ok_tensors = [load_and_preprocess_image(img, preprocess_pipeline) for img in ok_image_paths]
        junk_tensors = [load_and_preprocess_image(img, preprocess_pipeline) for img in junk_image_paths]

        good_features_list = [extract_image_features(model, tensor) for tensor in good_tensors]
        ok_features_list = [extract_image_features(model, tensor) for tensor in ok_tensors]
        junk_features_list = [extract_image_features(model, tensor) for tensor in junk_tensors]


        similarities_good = compute_cosine_similarity(query_features, good_features_list)


        similarities_ok = compute_cosine_similarity(query_features, ok_features_list)


        similarities_junk = compute_cosine_similarity(query_features, junk_features_list)


        all_similarities = similarities_good + similarities_ok + similarities_junk
        relevant_images = [1] * len(similarities_good) + [0.75] * len(similarities_ok) + [0] * len(similarities_junk)
        relevant_images_hard = [0] * len(similarities_good) + [0.25] * len(similarities_ok) + [1] * len(similarities_junk)

        mAP = compute_mAP(all_similarities, relevant_images, len(similarities_good) + 0.75*len(similarities_ok))


        H = compute_mAP(all_similarities,  relevant_images_hard, len(similarities_junk) + 0.25*len(similarities_ok))


        print(f"Group: {group_name}, mAP: {mAP}, H: {H}")


        total_mAP += mAP
        total_H += H
        num_queries += 1


    average_mAP = total_mAP / num_queries
    average_H = total_H / num_queries


    print(f"Average mAP: {average_mAP}")
    print(f"Average H: {average_H}")


def extract_group_names(groundtruth_dir):
    group_names = set()
    for filename in os.listdir(groundtruth_dir):
        if filename.endswith('_query.txt'):
            group_name = filename.replace('_query.txt', '')
            group_names.add(group_name)
    return sorted(list(group_names))
```

```python

groundtruth_dir = 'paris_groundtruth/'


dataset_dir = 'paris/paris/'


preprocess_pipeline = get_preprocessing_pipeline()
model = dinov2_vits14


process_ground_truth(groundtruth_dir, dataset_dir, model, preprocess_pipeline)
```
